```python
import os
import pandas as pd

processed_dir = '../data/processed/filtered_files'
csv_files = [f for f in os.listdir(processed_dir) if f.endswith('.csv')]

dataframes = {}
for file in csv_files:
    df_name = os.path.splitext(file)[0]
    df_name = "df_" + df_name.split("_")[1]
    df_path = os.path.join(processed_dir, file)
    dataframes[df_name] = pd.read_csv(df_path, encoding="utf-8")
# Display the names of the loaded DataFrames
```

```python
# Ensure that df_distances is available from the existing dataframes dictionary.
df_distances = dataframes.get("df_distances")
df_municipalities = dataframes.get("df_municipalities")
df_schools = dataframes.get("df_schools")
df_industry = dataframes.get("df_industry")
df_trains = dataframes.get("df_trains")
df_connectivity = dataframes.get("df_connectivity")
df_climate = dataframes.get("df_climate")
df_airports = dataframes.get("df_airports")
df_hospitals = dataframes.get("df_hospitals")
df_demographics = dataframes.get("df_demographics")
df_regional_trains = dataframes.get("df_regionaltrains")
```
